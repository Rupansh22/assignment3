```python

# COMP2200/COMP6200 Assignment 3  #
# Directors Network Analysis      #


import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Load data
company_directorships = pd.read_csv("company_directorships.csv")
director_details = pd.read_csv("director-details.csv")

# Clean names for both files
company_directorships['director_name_clean'] = company_directorships['director_name'].str.strip().str.lower()
director_details['name_clean'] = director_details['name'].str.strip().str.lower()

# Build the network using CLEANED names from the start!
G = nx.Graph()

for company, group in company_directorships.groupby("cikcode"):
    directors = group["director_name_clean"].tolist()
    for i in range(len(directors)):
        for j in range(i+1, len(directors)):
            G.add_edge(directors[i], directors[j], company=company)

# Calculate centrality measures
degree_centrality = nx.degree_centrality(G)
eigenvector_centrality = nx.eigenvector_centrality(G, max_iter=1000)
betweenness_centrality = nx.betweenness_centrality(G)

# Build dataframe with cleaned director names
centrality_df = pd.DataFrame({
    'director_name_clean': list(degree_centrality.keys()),
    'Degree': list(degree_centrality.values()),
    'Eigenvector': list(eigenvector_centrality.values()),
    'Betweenness': list(betweenness_centrality.values())
})

# Merge successfully on cleaned names
final_df = pd.merge(centrality_df, director_details, on='director_name_clean', how='left')

# Display first few rows to verify
print(final_df.head())

# Plot Age Distribution
plt.figure(figsize=(8,5))
plt.hist(final_df['age'].dropna(), bins=20)
plt.title("Distribution of Director Ages")
plt.xlabel("Age")
plt.ylabel("Number of Directors")

plt.show()

# Plot Degree Centrality vs Compensation
plt.figure(figsize=(8,5))
plt.scatter(final_df['Degree'], final_df['compensation'], alpha=0.6)
plt.title("Degree Centrality vs Compensation")
plt.xlabel("Degree Centrality")
plt.ylabel("Compensation ($)")

plt.show()
```

Task 1: Centrality Measures – Business Context
The board network analysis employed different centrality measures to determine which directors could assist the venture capital firm in selling its portfolio company.

Degree Centrality:
The number of directors who directly connect to each individual determines their degree centrality score. Directors who have higher degree centrality scores participate in multiple boards that contain different individuals thus expanding their professional connections. A director who maintains relationships with numerous people will have better opportunities to establish connections with potential acquirers.

Eigenvector Centrality:
Eigenvector centrality extends the basic measure by evaluating both the contacts of a director and the level of influence their connections possess. A director who achieves high eigenvector scores maintains relationships with influential people in their network. Directors with high betweenness centrality possess stronger network influence which enables them to reach essential decision-makers at companies that could execute acquisitions.

Betweenness Centrality:
Betweenness centrality identifies the directors who link different isolated groups together. These individuals link different sections of the network which otherwise would not communicate. Directors with high betweenness centrality serve as vital connectors between the venture capital-backed company and businesses it would not normally reach.

```python
!pip install rapidfuzz

from rapidfuzz import process

# Build a mapping from company_directorships to director_details using fuzzy matching
name_map = {}

for director_name in company_directorships['director_name_clean'].unique():
    match, score, _ = process.extractOne(director_name, director_details['name_clean'])
    if score >= 90:  # adjust threshold as needed
        name_map[director_name] = match

# Apply mapping
centrality_df['matched_name'] = centrality_df['director_name_clean'].map(name_map)

# Merge on matched names
final_df = pd.merge(
    centrality_df,
    director_details,
    left_on='matched_name',
    right_on='name_clean',
    how='left'
)

# Now print and check:
print(final_df[['director_name_clean','matched_name','name','age']].head())
```

```python
# Age Distribution Plot

# Make sure age column is numeric
final_df['age'] = pd.to_numeric(final_df['age'], errors='coerce')

# Drop missing values
age_data = final_df['age'].dropna()

plt.figure(figsize=(8,5))
plt.hist(age_data, bins=20, color='skyblue', edgecolor='black')
plt.title("Distribution of Director Ages")
plt.xlabel("Age")
plt.ylabel("Number of Directors")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
